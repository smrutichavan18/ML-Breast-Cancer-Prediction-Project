```python
bcp['diagnosis'] = bcp['diagnosis'].map({'M':1, 'B':0})
```

```python
bcp.head(10)
```

```python
X = bcp.drop('diagnosis', axis=1)
X.head(3)
```

```python
y = bcp['diagnosis']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
sts = StandardScaler()
X_train = sts.fit_transform(X_train)
X_test = sts.transform(X_test)
```

```python
X_train
```

# MACHINE LEARNING


# Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
lr_acc = accuracy_score(y_test, y_pred)
print(lr_acc)
```

```python
comp_results = pd.DataFrame()
comp_results
```

```python
tempResults = pd.DataFrame({'Algorithm':['Logistic Regression Method'], 'Accuracy':[lr_acc]})
comp_results = pd.concat( [comp_results, tempResults] )
comp_results = comp_results[['Algorithm','Accuracy']]
comp_results
```

# DECISION TREE CLASSIFIER 

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
dtc_acc = accuracy_score(y_test, y_pred)
print(dtc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Decision Tree Classifier'], 'Accuracy':[dtc_acc]})
comp_results = pd.concat( [comp_results, tempResults] )
comp_results = comp_results[['Algorithm','Accuracy']]
comp_results
```

# RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
rfc_acc = accuracy_score(y_test, y_pred)
print(rfc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Random Forest Classifier'], 'Accuracy':[rfc_acc]})
comp_results = pd.concat( [comp_results, tempResults] )
comp_results = comp_results[['Algorithm','Accuracy']]
comp_results
```
